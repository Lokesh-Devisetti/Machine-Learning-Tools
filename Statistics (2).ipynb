**Outliers**

```python
a = [1,2,2,2,3,3,4,5,5,5,6,6,6,6,7,8,8,9,27]

b = pd.DataFrame({'x' : a})

b
```

```python
b['x'].median()
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (5,3))

sns.boxplot(data = b['x'] , orient = 'h')
```

```python
# How to handle them

IQR = b['x'].quantile(0.75) - b['x'].quantile(0.25)

print(IQR)

lower = b['x'].quantile(0.25) - 1.5 * (IQR)
upper = b['x'].quantile(0.75) + 1.5 * (IQR)

print(lower)
print(upper)
```

```python
import numpy as np
```

```python
b['x_c'] = np.where(b['x'] > upper, upper,
                    np.where(b['x'] < lower ,lower , b['x']))
```

```python
b
```

```python
# before
plt.figure(figsize = (5,3))
sns.boxplot(b['x'] , orient = 'h')
```

```python
#after
plt.figure(figsize = (5,3))
sns.boxplot(b['x_c'],orient = 'h')
```

**variance**

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'Rohit':[2,18,64,89,12,0] , 'Kohli':[28,31,37,39,41,30]})
df
```

```python
df['Rohit'].var()
```

```python
df['Rohit'].mean()
```

```python
# using pure python implementing variance concept on rohit column

c = 0
d = 0


for i in df['Rohit']:
  c = c + i

mean_ = c / len(df)

print('mean of the column : ',mean_)


for j in df['Rohit']:
  d = d + (j - mean_)**2

print(f'the variance of the rohit : ',d / (len(df) - 1))
```

```python
print(df['Kohli'].mean())
print(df['Kohli'].var())
```

```python
# using pure python implementing variance concept on rohit column

c = 0
d = 0


for i in df['Kohli']:
  c = c + i

mean_ = c / len(df)

print('mean of the column : ',mean_)


for j in df['Kohli']:
  d = d + (j - mean_)**2

print(f'the variance of the Kohli : ',d / (len(df) - 1))
```

**standard deviation**

```python
# rohit

df['Rohit'].std()
```

```python
# using pure python implementing standard devitation concept on rohit column

c = 0
d = 0


for i in df['Rohit']:
  c = c + i

mean_ = c / len(df)

print('mean of the column : ',mean_)


for j in df['Rohit']:
  d = d + (j - mean_)**2

print(f'the std of the Rohit : ',np.sqrt(d / (len(df) - 1)))
```
