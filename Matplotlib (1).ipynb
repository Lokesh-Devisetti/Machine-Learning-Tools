**Matplotlib**

Matplotlib is used for representing the data in the form of images..

Basically EDA -> Exploratory Data Analaysis means visualy showing the data...

ex.EDA. exsheet -> graphical representation manner

There are few plotting techniques which we are going to learn in matplotlib

- `scatter plot`
- `line plot`
- `Legend Function`
- `Historgram`
- `Pie chart`
- `subplots`
- `barcharts`

**Scatter plot**

scatter works with x_axis and y_axis where based on x_axis we are going to plot y_axis .. point ...

```python
import matplotlib.pyplot as plt
```

```python
# take 2d Data and represent them

x = [2,4,7,9,11]
y = [1,4,3,12,6]

plt.scatter(x , y) # scatter plot

plt.show()   # show the image infront of you..
```

```python
# take 2nd data and represent them
plt.figure(figsize=(5,3))
x = [2,4,7,9,11]
y = [1,4,3,12,6]
plt.title('scatter plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.xticks([])
plt.yticks([])
plt.scatter(x,y,color='r',marker='d')
plt.show()
```

**Line_plot**

Just connecting the Scatter_plot points

```python
# take 2d Data and represent them

plt.figure(figsize = (5,3))

x = [2,4,7,9,11]
y = [1,4,3,12,6]

plt.title('scatter plot')
plt.xlabel('X_axis')
plt.ylabel('y_axis')

plt.xticks([])
plt.yticks([])


plt.plot(x , y , color = 'b' , marker = 's') # line plot

plt.show()   # show the image infront of you..
```


**Legend Functions**

this function basically used to plot the data and give some identity for each plot -> so that we can easily recognize the plot..

```python
days=[1,2,3,4,5,6,7]
a=[50,34,54,89,42,86,43]
b=[43,42,40,44,33,35,37]
c=[45,48,48,46,40,42,41]

plt.figure(figsize = (5,2))

plt.plot(days,a) # days on x_axis , a on y_axis
plt.plot(days,b) # days on x_axis , b on y_axis
plt.plot(days,c) # days on x_axis , c on y_axis


plt.show()
```

```python
days=[1,2,3,4,5,6,7]
a=[50,34,54,89,42,86,43]
b=[43,42,40,44,33,35,37]
c=[45,48,48,46,40,42,41]
plt.figure(figsize = (5,2))

plt.plot(days,a ,label = 'days_a')  # days on x_axis , a on y_axis
plt.plot(days,b , label = 'days_b') # days on x_axis , b on y_axis
plt.plot(days,c , label = 'days_c') # days on x_axis , c on y_axis

plt.legend(loc = 1 , fontsize = 'medium')
plt.show()
```

**Histogram**

Histogram -> it finds the data which is present in between the given range

```python
Age = [3,5,9,10,14,18,19,20,23,27,34,35,36,38,42,45]

plt.hist(x = Age , bins = 5)

plt.show()
```

**subplots**

```python
plt.figure(figsize = (8,3))

days=[1,2,3,4,5,6,7]
a=[50,51,52,48,47,49,46]
b=[43,42,40,44,33,35,37]
c=[45,48,48,46,40,42,41]


# plot days and a:
plt.subplot(1,3,1)
plt.title('days_a')
plt.plot(days,a)

#plot days and b
plt.subplot(1,3,2)
plt.title('days_b')
plt.plot(days,b)

# plot days and c
plt.subplot(1,3,3)
plt.title('days_c')
plt.plot(days,c)

plt.show()
```

```python
plt.figure(figsize = (8,3))

days=[1,2,3,4,5,6,7]
a=[50,51,52,48,47,49,46]
b=[43,42,40,44,33,35,37]
c=[45,48,48,46,40,42,41]


# plot days and a:
plt.subplot(2,2,1)
plt.title('days_a')
plt.plot(days,a)


plt.subplot(2,2,2)
plt.title('days_b')
plt.plot(days,b)


plt.subplot(2,2,3)
plt.title('days_c')
plt.plot(days,c)

plt.show()
```

**Barchart**

Based on values we are going to keep the Bar

```python
Language = ['Python', 'Java', 'c++', '.NET', 'Go']
Profits = [1600 , 1000 , 400 , 200 , 50]

plt.bar(Language,Profits,color='g')

plt.show()
```


**Pie_chart**

Pie_chart -> Based on the values we are going to make a split the circle..

```python
profits = [60 , 40 , 30 , 10 , 1]

names = ['Avenger' , 'Avatar' , 'SR' , 'abc' , 'shark']

plt.pie(profits , labels = names , shadow = True , autopct = '%1.1f%%')

plt.show()
```

```python
profits = [60 , 40 , 30 , 10 , 1]

names = ['Avenger' , 'Avatar' , 'SR' , 'abc' , 'shark']

plt.pie(profits , labels = names , shadow = True , autopct = '%1.1f%%' , explode=[0 , 0 , 0.2 , 0.4 , 0])

plt.show()
```
